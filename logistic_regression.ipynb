Logistic Regression


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
dataset=pd.read_csv('framingham.csv')
dataset
```

```python
dataset.drop(dataset.columns[[0,5,7,8,9,10,11,12,13,14,15]],axis=1,inplace=True)
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="mean")
dataset['education']=imp.fit_transform(dataset[['education']])
dataset['cigsPerDay']=imp.fit_transform(dataset[['cigsPerDay']])
dataset.isnull().sum()
```

Creating feature matrix and dependent variable vector


```python
x=dataset.iloc[:,2:4].values
y=dataset.iloc[:,4].values
```

Feature Matrix

```python
x
```

Dependent Variable

```python
y
```

Splitting data into training data and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train.shape
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

Training the model

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
```

Prediction

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
