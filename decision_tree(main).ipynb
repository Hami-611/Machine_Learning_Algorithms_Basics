```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
dataset=pd.read_csv(r"/content/Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

Consider Statistical Columns

```python
X= dataset.iloc[:,2:3].values
y= dataset.iloc[:,4].values
```

```python
X[0]
```

```python
y[0]
```

```python
sns.heatmap(dataset.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier( random_state=6)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of the model: {}%".format(accuracy_score(y_test,y_pred)*100))
```

```python
accuracy = []
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

for i in range(1,10):
  model = DecisionTreeClassifier(max_depth = i, random_state=0)
  model.fit(X_train,y_train)
  pred = model.predict(X_test)
  score = accuracy_score(y_test,pred)
  accuracy.append(score)

plt.figure(figsize=(12,6))
plt.plot(range(1,10),accuracy,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title("Best Max_Depth")
plt.xlabel('pred')
plt.ylabel('score')
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = i, random_state=0)
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, pred))
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(max_leaf_nodes = 5, random_state=0)
model1.fit(X_train,y_train)
model1_pred = model1.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model1,filled=True)
plt.show()
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, model1_pred))
```
